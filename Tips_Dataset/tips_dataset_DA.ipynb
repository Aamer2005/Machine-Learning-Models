`Visualization `

```python
sns.histplot(dataset["total_bill"],kde=True)
plt.axvline(dataset["total_bill"].mean(),color = "red")
plt.annotate("mean",xy=(dataset["total_bill"].mean(),45),xytext=(25,46),arrowprops=dict(facecolor="black",shrink = 0.001))
```

```python
sns.histplot(dataset["tip"],kde=True,color="green")
plt.axvline(dataset["tip"].mean(),color = "red")
plt.annotate("mean",xy=(dataset["tip"].mean(),45),xytext=(4,50),arrowprops=dict(facecolor="blue",shrink = 0.001))
```
